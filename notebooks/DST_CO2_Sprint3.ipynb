### next apply feature selection methods such as PCA and wrapper method in order to reduce the  number of future of the model. the target is co2[g/km].

```python
from sklearn.decomposition import PCA
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Defining the features and target variable
X = data.drop(['CO2 (g/km)'], axis=1)
y = data['CO2 (g/km)']

# PCA
pca = PCA(n_components=0.95)  # Retain 95% of the variance
X_pca = pca.fit_transform(X)

# Wrapper Method: RFE with Linear Regression
# Note: Due to computational constraints, we'll limit the number of features to select for demonstration purposes
linear_reg = LinearRegression()
rfe = RFE(estimator=linear_reg, n_features_to_select=15, step=10)       # Selecting top 15 features
X_rfe = rfe.fit_transform(X, y)

# Information about the reduced datasets
pca_n_components = X_pca.shape[1]
rfe_selected_features = [X.columns[i] for i in range(len(X.columns)) if rfe.support_[i]]

# Creation of a DataFrame to visualize the feature selected by the RFE
rfe_selected_features_df = pd.DataFrame(rfe_selected_features)
```

### Given the computational demands of RFE with such a large dataset, a more straightforward and less computationally intensive method for feature selection is to use feature importance from a model like a decision tree or to examine the correlation coefficients between features and the target variable. This approach can provide insights into which features are most predictive of the target variable without the need for exhaustive computations.

```python
from sklearn.tree import DecisionTreeRegressor

# Instantiating and fitting the Decision Tree model
tree_model = DecisionTreeRegressor(random_state=0)
tree_model.fit(X, y)

# Getting feature importances
feature_importances = tree_model.feature_importances_

# Creation of a DataFrame to visualize the feature importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})
# Sorting the DataFrame by importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Displaying the top 15 most important features
feature_importance_df.head(15)
```

###  Elaboration of the first modeling iterations with baseline models and Analysis of initial results and assessing the relevance of your approach and model.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# Selecting the top 22 features for modeling
top_22_features = feature_importance_df['Feature'].head(22)
X_selected = X[top_22_features]

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Initialize models
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree Regressor": DecisionTreeRegressor(random_state=42),
    "Random Forest Regressor": RandomForestRegressor(n_estimators=100, random_state=42)
}

# Function to train and evaluate models

def train_evaluate(models, X_train, X_test, y_train, y_test):
    results = {}
    for name, model in models.items():
        # Training the model
        model.fit(X_train, y_train)
        # Predicting on the testing set
        y_pred = model.predict(X_test)
        # Evaluating the model
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        results[name] = {'MAE': mae, 'MSE': mse, 'RÂ² Score': r2}
    return results

# Training and evaluating the models
results = train_evaluate(models, X_train, X_test, y_train, y_test)

# Storing the results into dataframe

results_df = pd.DataFrame(results).T
```

### To deepen our analysis of the initial results and assess the relevance of the approach and models, we'll visualize the accuracy (RÂ² Score) and calculate the Root Mean Squared Error (RMSE) for the three baseline models: Linear Regression, Decision Tree Regressor, and Random Forest Regressor.

```python
# Calculating RMSE for each model and adding it to the results DataFrame
results_df['RMSE'] = results_df['MSE'].apply(np.sqrt)

# Visualization of model performance
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# RÂ² Scores
sns.barplot(x=results_df.index, y="RÂ² Score", data=results_df, ax=ax[0])
ax[0].set_title('RÂ² Score of Models')
ax[0].set_ylabel('RÂ² Score')
ax[0].set_ylim(0.95, 1)
for p in ax[0].patches:
    ax[0].annotate(format(p.get_height(), '.4f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

# RMSE
sns.barplot(x=results_df.index, y="RMSE", data=results_df, ax=ax[1])
ax[1].set_title('RMSE of Models')
ax[1].set_ylabel('RMSE')
for p in ax[1].patches:
    ax[1].annotate(format(p.get_height(), '.6f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

plt.tight_layout()
plt.show()

results_df[['RÂ² Score', 'RMSE']]
```

### 1. Gradient Boosting Machines (GBM): An ensemble technique that builds trees sequentially, with each tree trying to correct the errors of the previous one.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
import xgboost as xgb
import lightgbm as lgb

# Define a small grid for Random Forest hyperparameter tuning (for demonstration purposes)
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10],
}

# Grid Search for Random Forest
rf_grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=3, scoring='neg_mean_squared_error')
rf_grid_search.fit(X_train, y_train)

# Initialize and train other complex models
gbm = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gbm.fit(X_train, y_train)

xgboost_model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
xgboost_model.fit(X_train, y_train)

lightgbm_model = lgb.LGBMRegressor(n_estimators=100, learning_rate=0.1, max_depth=-1, random_state=42)
lightgbm_model.fit(X_train, y_train)

# Evaluate all models
models_optimized = {
    "Random Forest Optimized": rf_grid_search.best_estimator_,
    "Gradient Boosting Machines": gbm,
    "XGBoost": xgboost_model,
    "LightGBM": lightgbm_model
}

# Reuse the train_evaluate function defined earlier for evaluating models
results_optimized = train_evaluate(models_optimized, X_train, X_test, y_train, y_test)
results_optimized_df = pd.DataFrame(results_optimized).T
results_optimized_df['RMSE'] = results_optimized_df['MSE'].apply(np.sqrt)

results_optimized_df[['RÂ² Score', 'RMSE']]
```

### 1. Bagging and Boosting: 
### a) Bagging (Bootstrap Aggregating) improves model accuracy by training multiple models (usually of the same type) on different subsets of the training dataset, then averaging their predictions. A typical example is the Random Forest algorithm. 
### b) Boosting refers to a group of algorithms that convert weak learners to strong learners by focusing on mistakes of prior models. Boosting algorithms like Gradient Boosting, XGBoost, and LightGBM have been highly successful for various regression and classification problems.


### 2. Deep Learning. 
### We can use a Neural Network with several hidden layers. Deep learning can model complex nonlinear relationships and interactions between features but requires more data and computational resources.
### 3. Interpret Results 
### Using the chosen interpretability tools, we can assess which features are most influential for predictions and understand how the model makes its decisions, enhancing trust in the model's predictions.

```python
# Refining XGBoost with Hyperparameter Tuning.

from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV

# Assuming X_train and y_train are already defined using the reduced dataset

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1]
}

# Initialize the XGBoost model
xgb_model = XGBRegressor()

# Perform grid search
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Best parameters and model
best_params = grid_search.best_params_
optimized_xgb = grid_search.best_estimator_
```
